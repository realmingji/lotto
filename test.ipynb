```python
import numpy as np
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
df['회차'] = range(1, len(df)+1)
```

```python
predict_numbers = {}  # 다음 회차 예측
```

```python
for col in ['1','2','3','4','5','6','보너스']:
  model = ARIMA(df[col], order=(5,1,0))
  model_fit = model.fit()
  
  predicted = model_fit.forecast()
  predict_numbers[col] = np.round(predicted.iloc[0])
```

```python
predict_numbers
```

```python
import tkinter as tk
from tkinter import messagebox
```

```python
def predict_lotto_numbers():
  try:
    csv_file_path='selected_data.csv'
    df = pd.read_csv(csv_file_path, encoding='cp949')
    df['회차'] = range(1, len(df)+1)
    predict_numbers = {}
    for col in ['1','2','3','4','5','6','보너스']:
      X = df.drop(columns=[col])
      y = df[col]
      X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
      model = RandomForestRegressor(n_estimators=100, random_state=42)
      model.fit(X_train, y_train)
      predicted = model.predict(X_test.iloc[-1].values.reshape(1, -1))
      predict_numbers[col] = np.round(predicted[0])
      predicted_text = "\n".join([f"{k}: {v}" for k, v in predict_numbers.items()])
      messagebox.showinfo("예측 결과", predicted_text)
      
  except Exception as e:
    messagebox.showerror("오류", str(e))
```

```python
root = tk.Tk()
root.title("로또 번호 예측기")
predicted_button = tk.Button(root, text="번호 예측하기", command=predict_lotto_numbers)
predicted_button.pack(pady=20)

root.mainloop()
```
